```python
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt
from tensorflow.keras.preprocessing.image import ImageDataGenerator, load_img, img_to_array
```

```python
#loading the image
path = "./Images/dog.jpg"
original_image = load_img(path, target_size=(150,150))# Resize for consistency
image_array = img_to_array(original_image) #Image array shape(150,150,3)
image_array = image_array.reshape((1,) +image_array.shape) # (batch_size, height, width, channels).
```

```python
datagen = ImageDataGenerator(
    rotation_range=30,       # Random rotation between -30 and 30 degrees
    width_shift_range=0.2,   # Horizontal shift
    height_shift_range=0.2,  # Vertical shift
    shear_range=0.2,         # Shear transformation
    zoom_range=0.2,          # Random zoom
    horizontal_flip=True,    # Horizontal flipping
    fill_mode='nearest'      # Filling strategy for new pixels
)
```

```python
# Generate and display augmented images
augmented_images = datagen.flow(image_array, batch_size=1)

# Visualize augmented images
plt.figure(figsize=(12, 6))
for i in range(10):  
    plt.subplot(2, 5, i + 1)
    batch = next(augmented_images)
    augmented_image = batch[0].astype('uint8')
    plt.imshow(augmented_image)
  

plt.suptitle("Augmented Dog Images")
plt.tight_layout()
plt.show()
```
